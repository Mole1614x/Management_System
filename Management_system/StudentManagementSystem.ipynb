```python
####################### Student Management System ###############################

student = {}

while True:
    
    print("########### Student Management System #################")
    print("1.Add Student")
    print("2. Display")
    print("3. Search")
    print("4. Update")
    print("5. Delete")
    print("6. Exit")
    
    ch = int(input("Enter your choice:"))
    
    if ch == 1:
        
        roll_no = int(input("Enter your roll no :"))
        name = input("Enter your name :").strip().title()
        marks = int(input("Enter your marks :"))
        
        student[roll_no] = {"Name": name, "Marks": marks}
        
        print(f" The Student data has been updated: {student}")
        
    elif ch == 2:
        
        for k,v in student.items():
            print(f"{k} : {v}")
            
    elif ch == 3:
        
        roll_no = int(input("Enter your roll no :"))
        
        if roll_no not in student:
            print("Student not found")
        else:
            print(f"The searched student id: {student[roll_no]}")
    
    elif ch == 4:
        
        roll_no = int(input("Enter your roll no :"))
        update_marks = int(input("Enter your new marks :"))
        
        
        if roll_no not in student:
            print("Student not found")
        else:
            student[roll_no]["Marks"] = update_marks
            print(f"The updated marks of student is: {student[roll_no]}")
    
    elif ch == 5:
        
        del_roll_no = int(input("Enter your roll no :"))
        
        if del_roll_no not in student:
            print("Student not found")
        else:
            del student[del_roll_no]         
            print(f"The updated student is: {student}") 
            
    elif ch == 6:
        break
        
    else:
        print("Invalid Choice")
        
```

```python
################### Simple Employee System ##################

emp = {}

while True:
    
    print("########### Employee System #################")
    print("1. Add ")
    print("2. Display")
    print("3. Search")
    print("4. Delete")
    print("5. Exit")
    
    ch = int(input("Enter your choice:"))
    
    if ch == 1:
        
        e_id = int(input("Enter your Employee id:"))
        name = input("Enter your name:")
        salary = int(input("Enter your salary:"))
        
        emp[e_id] = {"Name" :name, "Salary" : salary}
        
        print(emp)
    
    elif ch == 2:
        
        for k,v in emp.items():
            print(f"{k} --> {v}")
            
    elif ch == 3:                         ####### highest salary
    
        highest_emp = max(emp.items(), key=lambda x: x[1]["Salary"])

        print("Employee ID:", highest_emp[0])
        print("Details:", highest_emp[1])
        
        
    elif ch == 4:
                
        e_id = int(input("Enter your Employee id:"))
        
        if e_id not in emp:
            print("Employee not Found")
        else:
            del emp[e_id]
            print(f"{emp}")
            
    elif ch == 5:
        
        print("Exiting...")
        break
        
    else:
        print("Invalid Choice")
```

```python

################ LIBRARY MANAGEMENT SYSTEM ##################

book = {}

class lms:
    
    def __init__(self, title, author, price, qty):
        self.title = title
        self.author = author
        self.price = price
        self.qty = qty
        
        # FIXED 1: Store book in dictionary inside constructor
        book[self.title] = {
            "Author": self.author,
            "Price": self.price,
            "Quantity": self.qty
        }
    
    def __str__(self):
        return f"|Title: {self.title}| Author: {self.author}| Price: {self.price}| Quantity: {self.qty}|"
    
    def add_book(self):
        # FIXED 2: No KeyError now because constructor already stores data
        print("Book added successfully!")
        print(book[self.title])
    
    def issue_book(self):
        
        title = input("Enter Title: ")
        
        if title not in book:
            print("The book does not exist!")
        else:
            if book[title]["Quantity"] > 0:
                book[title]["Quantity"] -= 1
                print("Book issued successfully!")
            else:
                print("Book not available!")
    
    def return_book(self):
        
        title = input("Enter Title: ")
        
        if title not in book:
            print("The book does not exist!")  
        else:
            book[title]["Quantity"] += 1
            print("Book is now available")
    
    def display(self):
        
        if not book:   # FIXED 3: handle empty dictionary case
            print("No books available!")
        else:
            for k, v in book.items():
                print(f"{k} : {v}")
    
    def search_book(self):
        
        title = input("Enter Title: ")
        
        if title not in book:
            print("The book does not exist!")
        else:
            print(f"Book has been found: {book[title]}")


# ================= MAIN LOOP =================

while True:
    
    print("\n######### LIBRARY MANAGEMENT SYSTEM ################")
    print("1. ADD Book") 
    print("2. Issue Book")
    print("3. Return Book")
    print("4. Display")
    print("5. Search")
    print("6. Exit")
    
    ch = int(input("Enter your choice: "))

    if ch == 1:
        title = input("Enter Title: ")
        author = input("Enter Author: ")
        price = int(input("Enter Price: "))
        qty = int(input("Enter qty: "))
        
        l1 = lms(title, author, price, qty)
        print("Book object created successfully!")   # FIXED 4: clarity message

    elif ch == 2:
        l1.issue_book()

    elif ch == 3:
        l1.return_book()

    elif ch == 4:
        l1.display()

    elif ch == 5:
        l1.search_book()

    elif ch == 6:
        print("Exiting...")
        break

    else:
        print("Invalid Choice!")
```
