```python
import sqlite3

# connect/create database
conn = sqlite3.connect("student.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER,
    marks REAL
)
""")

# Insert 5 records (NO id here)
students_data = [
    ("Anmol", 23, 85),
    ("Avi", 23, 87),
    ("Sneh", 24, 81),
    ("Priyanshu", 22, 90),
    ("Farhan", 26, 80)
]

cursor.executemany(
    "INSERT INTO students(name, age, marks) VALUES (?,?,?)",
    students_data
)

# Fetch all records
cursor.execute("SELECT * FROM students")
records = cursor.fetchall()

print("Student Records:")
for row in records:
    print(row)

conn.commit()
conn.close()
```

```python
import sqlite3

# ================= LOGIN SYSTEM =================

def register():
    username = input("Enter new username: ")
    password = input("Enter new password: ")

    with open("users.txt", "a") as f:
        f.write(username + "," + password + "\n")

    print("Registered successfully!")


def login():
    username = input("Enter username: ")
    password = input("Enter password: ")

    try:
        with open("users.txt", "r") as f:
            users = f.readlines()
    except FileNotFoundError:
        print("No users found. Please register first.")
        return False

    for user in users:
        u, p = user.strip().split(",")

        if u == username and p == password:
            print("Login successful!")
            return True

    print("Invalid credentials")
    return False


def login_system():
    attempts = 3

    while attempts > 0:
        if login():
            return True
        else:
            attempts -= 1
            print(f"Attempts left: {attempts}")

    print("Too many failed attempts!")
    return False

# ================= DATABASE =================

conn = sqlite3.connect("price.db", isolation_level = None)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS ladies(
id integer primary key autoincrement,
name text, 
dalal text,
per_night integer,
rate real
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS issued(
id INTEGER,
name TEXT,
dalal TEXT,
per_night INTEGER,
rate REAL
)
""")

# Clear table
cursor.execute("DELETE FROM ladies")

# Reset AUTOINCREMENT counter
cursor.execute("DELETE FROM sqlite_sequence WHERE name='ladies'")


products = [
    ("jisse pyaar ho jaata h", "not found", 150, 2999),
    ("jo nhi maanti", "mama ka ladka", 10, 299),
    ("chamach se khaane waali", "koi bahar ka", 300, 5999),
    ("Laad", "khud dalal h", 200, 3999),
    ("didi mein didi", "service-needed", 0, 99),
    ("Legion","jispe ipad hai woh dalal h", 499 , 4999),
    ("Gyaan ki devi", "attitude boy", 159, 1599),
    ("Flat Chested 1", "Flat chested 2", 99, 99),
    ("Flat Chested 2", "Flat Chested 1", 99, 99),
    ("Bonnie", "Babu Iron mutthi dalal",20, 200 ),
    ("Nathan ka maal", "nsnathan", 59, 599),
    ("milf","didi dalal",799,7999),
    ("petrol","access 125",659,6599),
    ("D-mart","kalue yeh toh leke hi manega",459,4599),
    ("Behen","North Bhai lambi naak waala",259,2599),
    ("Chachi","Elder wand", 9, 199),
    ("AAO ka maal","AAO", 499,4999),
    ("Langdi Moti","Fetish waale dalal",699,4599),
    ("Hostel ka Item CR","9 ladke dalal", 0 , 49),
    ("Judawaa 1", "Judwaa 2", 699, 12999)
]

cursor.executemany("INSERT INTO ladies(name,dalal,per_night,rate) VALUES(?,?,?,?)",products)

# ================= FUNCTIONS =================
def add_product():

    ### INSERT NEW ENTRY

    name = input("Enter your name:")
    dalal = input("Dalal ka naam:")
    per_night = int(input("Enter per_night:"))
    rate = int(input("Enter rate of product:"))

    cursor.execute("insert into ladies(name,dalal,per_night,rate) values(?,?,?,?)",(name,dalal,per_night,rate))
    conn.commit()   # ✅ important
    print("Added successfully!")

def view_product():
    cursor.execute("SELECT * FROM ladies")
    
    print("\n Product details\n")
    for id, name, dalal, per_night, rate in cursor.fetchall():
        print(f"ID:{id} | Name:{name} | Dalal:{dalal} | Night:{per_night} | Rate:{rate}") 
    
def search_product(): 
    
    print("1. Search by name")
    print("2. Search by rate")
    
    choice = int(input("Enter Your choice:"))
    
    if choice == 1:
        
        name = input("Enter name to search: ")
        cursor.execute("SELECT * FROM ladies WHERE name LIKE ?",('%' + name + '%',))
        print("\nResult:", cursor.fetchall())
    
    elif choice == 2:
        
        rate = int(input("Enter minimum rate: "))
        cursor.execute("SELECT * FROM ladies WHERE rate >= ?", (rate,))   # comma make it a tuple
        print("\nResult:", cursor.fetchall())

    else:
        print("Invalid choice")
    
def sort_product():

    print("1. Sort in Asc")
    print("2. Sort in Dsc")

        
    choice = int(input("Enter Your choice:"))
    
    
    # Sorting Ascending
    
    if choice == 1:
        cursor.execute("""
        SELECT name,rate,per_night 
        FROM ladies 
        WHERE per_night >= 399 
        ORDER BY per_night ASC
        """)   # FIXED
        print("\nSorted ASC:", cursor.fetchall())

    # Sorting Descending
    elif choice == 2:
        
        cursor.execute("""
        SELECT name,rate,per_night 
        FROM ladies 
        WHERE per_night >= 399 
        ORDER BY per_night DESC
        """)   # FIXED

        print("\nSorted DESC:", cursor.fetchall())
    
    else:
        print("Invalid choice!")
        
def update_product():
  
    ##### update datails


    update_id = int(input("\nEnter ID to update: "))
    new_rate = float(input("Enter new rate: "))

    cursor.execute("update ladies set rate = ? where id = ?",(new_rate,update_id))
    print("Record updated successfully!") 
    
def delete_product():

    # delete

    delete_id = int(input("\nEnter ID to delete: "))

    cursor.execute(
        "DELETE FROM ladies WHERE id = ?",
        (delete_id,)
    )

    print("Record deleted!")


def Total():
    
    print("1. Total Revenue")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")
    print("5. Count")

        
    choice = int(input("Enter Your choice:"))
    
    ### Total Revenue Collection
    if choice == 1:
        cursor.execute("select sum(rate) from ladies")
        print(f"\nTotal Revenue: {cursor.fetchone()[0]}")
    
    # average rate (name cannot be added directly)
    elif choice == 2:
        cursor.execute("SELECT AVG(rate) FROM ladies")
        print("\nAverage:", cursor.fetchone()[0])

    elif choice == 3:
        
    # Maximum with name
        cursor.execute("""
        SELECT name, rate 
        FROM ladies 
        WHERE rate = (SELECT MAX(rate) FROM ladies)
        """)
        print("Max:", cursor.fetchall())

    # Minimum with name
    
    elif choice == 4:
        cursor.execute("""
        SELECT name, rate 
        FROM ladies 
        WHERE rate = (SELECT MIN(rate) FROM ladies)
        """)
        print("Min:", cursor.fetchall())
        
    elif choice == 5:
        
        # count (name cannot be added directly)
        cursor.execute("SELECT COUNT(*) FROM ladies")
        print("Count:", cursor.fetchone()[0])
    else:
        print("Invalid Choice!")
        
def premium():
    # Top Three Items

    cursor.execute("""
    SELECT name, rate 
    FROM ladies 
    ORDER BY rate DESC 
    LIMIT 3
    """)

    print("\nTop 3 Highest:")
    for row in cursor.fetchall():
        print(row)

def issue_product():
    issue_id = int(input("\nEnter ID to issue: "))

    cursor.execute("SELECT * FROM ladies WHERE id = ?", (issue_id,))
    product = cursor.fetchone()

    if product:
        # Insert into issued table
        cursor.execute("INSERT INTO issued VALUES(?,?,?,?,?)", product)

        # Delete from ladies table
        cursor.execute("DELETE FROM ladies WHERE id = ?", (issue_id,))

        conn.commit()
        print("Product issued and moved to issued list!")
    else:
        print("Invalid ID!")
        
def return_product():
    return_id = int(input("\nEnter ID to return: "))

    cursor.execute("SELECT * FROM issued WHERE id = ?", (return_id,))
    product = cursor.fetchone()

    if product:
        # Insert back into ladies
        cursor.execute("INSERT INTO ladies VALUES(?,?,?,?,?)", product)

        # Delete from issued table
        cursor.execute("DELETE FROM issued WHERE id = ?", (return_id,))

        conn.commit()
        print("Product returned and added back!")
    else:
        print("Invalid ID!")
        
        
# ================= LOGIN START =================

print("===== WELCOME TO FINEST LADIES.COM =====")
print("1. Register")
print("2. Login")

choice = int(input("Enter choice: "))

if choice == 1:
    register()

if not login_system():
    exit()

##### BIDDING LEFT    
    
    
# ================= MAIN MENU =================

while True:
    
    print("$------------------ Finest Ladies.Com ------------------------$")
    print("1. Add new product")
    print("2. View Available product")
    print("3. Search product")
    print("4. Sort product")
    print("5. Update product")
    print("6. Delete product")
    print("7. Revenue of product")
    print("8. Premium product")
    print("9. Issue product")
    print("10. Return product")
    print("11. Exit")
    
    ch = int(input("Enter your choice:"))
    
    if ch == 1:
        add_product()
    elif ch == 2:
        view_product()
    elif ch == 3:
        search_product()
    elif ch == 4:
        sort_product()
    elif ch == 5:
        update_product()
    elif ch == 6:
        delete_product()
    elif ch == 7:
        Total()
    elif ch == 8:
        premium()
    elif ch == 9:
        issue_product()
    elif ch == 10:
        return_product()
    elif ch == 11:
        print("Exiting...")
        break
    else:
        print("Invalid Choice!")
```
